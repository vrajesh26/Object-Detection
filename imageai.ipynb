```python
from imageai.Detection import ObjectDetection
import os
```

```python
execution_path = "C:/Users/HP/imageai"
```

```python
detector = ObjectDetection()
```

```python
detector.setModelTypeAsRetinaNet()
```

```python
detector.setModelPath("C:/Users/HP/resnet50_coco_best_v2.0.1.h5")
```

```python
detector.loadModel()
```

```python
execution_path
```

```python
detections = detector.detectObjectsFromImage(input_image="C:/Users/HP/imageai/car.jpg", output_image_path="C:/Users/HP/imageai/carnew.jpg")
```

```python
for eachObject in detections:
    print(eachObject["name"] , " : " , eachObject["percentage_probability"] )
```
